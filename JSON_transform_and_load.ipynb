```python
import json
from pymongo import MongoClient

client = MongoClient('localhost', 27017)
db = client['SD_housing']
collection_zips = db['zip_data']

db.zip_data.drop()


# Read file
with open('attom_output.txt') as json_file:  
    data = json.load(json_file)
    
    for i in range (0,len(data)):
        
        try:
            
            _id=collection_zips.insert(data[i]['response']['result']['package']['item'])[0]

            zip5=data[i]['response']['inputparameter']['AreaId'][2:]
            collection_zips.update_one({'_id': _id}, {"$set": {'zipcode': zip5}} )

        except KeyError:
            continue
            
client.close()
collection_zips.find_one()
```
